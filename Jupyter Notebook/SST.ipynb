```python
def BigData_reader(folder_path,val_name ='value' ,merge_col = ['lat','lon']):
    import geopandas as gpd
    import pandas as pd
    import os
    
    #read key
    key_path = folder_path + '/key.zip'
    key = gpd.read_file(key_path)
    
    # read value:Get a list of all the files in the directory
    files = os.listdir(folder_path)
    extension = '.feather'
    # Get a list of all the files in the directory with the specified extension
    value_files = [file for file in os.listdir(folder_path) if file.endswith(extension)]
    
    #merge key and value
    gdf = gpd.GeoDataFrame() 
    for v_f in value_files:
      value_path = folder_path+'/'+v_f
      value = pd.read_feather(value_path)
      gdf_temp = key.merge(value,on=merge_col)
      keep_col = [item for item in gdf_temp.columns if item not in merge_col]
      gdf_temp = gdf_temp[keep_col]
      gdf = pd.concat([gdf, gdf_temp], ignore_index=True)
    
    return gdf
```

```python
## Sea Surface Temperature in Norway from 2016 to 2022

BigData_reader('../data/SST_2016_22/')
```
